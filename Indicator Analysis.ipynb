```python
import matplotlib.pyplot as plt
import pandas_datareader.data as wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns


```

```python
# import data
start = '2016-01-01'
end = '2018-11-30'

aapl = wb.DataReader('AAPL', 'yahoo', start, end)
tsla = wb.DataReader('TSLA', 'yahoo', start, end)
gs = wb.DataReader('GS', 'yahoo', start, end)
ms = wb.DataReader('MS', 'yahoo', start, end)

df = wb.DataReader(['AAPL','TSLA','GS','MS'], 'yahoo', start, end)
df.info()
df.describe()
```

```python
# create a line plot showing Open price for these four stocks
tickers = ['AAPL', 'TSLA', 'MS', 'GS']
for tick in tickers:
    df['Open'][tick].plot(figsize=(12,4),label=tick)
plt.legend()
```

```python
# create a line plot showing Close price for these four stocks
tickers = ['AAPL', 'TSLA', 'MS', 'GS']
for tick in tickers:
    df['Close'][tick].plot(figsize=(12,4),label=tick)
plt.legend()
```

```python
# create a line plot showing Close price for these four stocks
tickers = ['AAPL', 'TSLA', 'MS', 'GS']
for tick in tickers:
    df['Volume'][tick].plot(figsize=(12,4),label=tick)
plt.legend()
```

```python
# create a line plot showing 20 days SMA  for these four stocks

tickers = ['AAPL', 'TSLA', 'MS', 'GS']

for tick in tickers:
   # df[f'ROC: {tick}'] = (df['Close'][tick] - df['Open']['GS'])*df['Volume'][tick]
    df['ROC',tick] = df.rolling(window=20).mean()['Close'][tick]

    df['ROC',tick].plot(figsize=(12,4),label=tick)
    
plt.legend()
```

```python
# create a line plot showing return of rate for these four stocks

tickers = ['AAPL', 'TSLA', 'MS', 'GS']

for tick in tickers:
    df['pct',tick] = df['Close'][tick].pct_change()
    df['pct',tick].plot(figsize=(12,4),label=tick)
    
plt.legend()
```
